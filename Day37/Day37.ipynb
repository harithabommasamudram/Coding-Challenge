```python
def get_marks():
    marks = []
    for i in range(3):
        m = int(input(f"Enter marks for subject {i+1}: "))
        marks.append(m)
    return marks
```

```python
def calculate_total(marks):
    return sum(marks)
```

```python
def calculate_average(total):
    return total / 3
```

```python
def get_grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 75:
        return "B"
    elif avg >= 50:
        return "C"
    else:
        return "D"
```

```python
# Main program

marks = get_marks()
total = calculate_total(marks)
average = calculate_average(total)
grade = get_grade(average)

print("Total Marks:", total)
print("Average Marks:", average)
print("Grade:", grade)
```
