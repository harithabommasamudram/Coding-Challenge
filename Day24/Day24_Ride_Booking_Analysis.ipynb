```python
rides = (
    (101, "Ravi", "Bangalore", 250, 4.5),
    (102, "Anjali", "Hyderabad", 180, 4.2),
    (103, "Kiran", "Bangalore", 300, 4.8),
    (104, "Meena", "Chennai", 220, 4.0),
    (105, "Arjun", "Bangalore", 150, 3.9)
)
```

```python
total_rides = len(rides)
print("Total Rides:", total_rides)
```

```python
print("Bangalore Rides:")

for ride in rides:
    if ride[2] == "Bangalore":
        print(ride)
```

```python
total_revenue = 0

for ride in rides:
    if ride[2] == "Bangalore":
        total_revenue += ride[3]

print("Total Bangalore Revenue:", total_revenue)
```

```python
highest_rated = max(rides, key=lambda x: x[4])
print("Highest Rated Ride:", highest_rated)
```

```python
rating_sum = {}
rating_count = {}

for ride in rides:
    city = ride[2]
    rating = ride[4]
    
    if city not in rating_sum:
        rating_sum[city] = rating
        rating_count[city] = 1
    else:
        rating_sum[city] += rating
        rating_count[city] += 1

# Calculate average
avg_rating = {}

for city in rating_sum:
    avg_rating[city] = round(rating_sum[city] / rating_count[city], 2)

print("Average Rating per City:")
print(avg_rating)
```
