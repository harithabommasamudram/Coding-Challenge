```python
salary = 50000
rent = 12000
food = 8000
transport = 3000
entertainment = 4000
```

```python
total_expenses = rent + food + transport + entertainment
print("Total Expenses:", total_expenses)
```

```python
savings = salary - total_expenses
print("Savings:", savings)
```

```python
print("Is rent greater than food?", rent > food)
print("Is transport less than entertainment?", transport < entertainment)
```

```python
if savings > 10000:
    print("Good Saving")
elif savings >= 5000 and savings <= 10000:
    print("Average Saving")
else:
    print("Low Saving")
```

```python
salary += salary * 0.10
print("Updated Salary:", salary)
```

```python
salary = 50000
rent = 12000
food = 8000
transport = 3000
entertainment = 4000

# Task 1
total_expenses = rent + food + transport + entertainment
print("Total Expenses:", total_expenses)

# Task 2
savings = salary - total_expenses
print("Savings:", savings)

# Task 3
print("Is rent greater than food?", rent > food)
print("Is transport less than entertainment?", transport < entertainment)

# Task 4
if savings > 10000:
    print("Good Saving")
elif savings >= 5000 and savings <= 10000:
    print("Average Saving")
else:
    print("Low Saving")

# Task 5
salary += salary * 0.10
print("Updated Salary:", salary)
```
