```python
students = ("sunitha", "RAVI", "aNu")
formatted_names = []
for name in students:
        formatted_names.append(name.capitalize())
formatted_names = tuple(formatted_names)
print(formatted_names)
```

```python
subjects = ("Math", "Science", "English", "Math", "Science")
total_subjects = len(subjects)
unique_subjects = set(subjects)
unique_count = len(unique_subjects)
print("Total:", total_subjects)
print("Unique:", unique_subjects)
print("Unique Count:", unique_count)
```

```python
feedback = ("  good course  ", " excellent ", "average  ")
cleaned_feedback = []
for fb in feedback:
        cleaned_feedback.append(fb.strip())
cleaned_feedback = tuple(cleaned_feedback)
print(cleaned_feedback)
```

```python
names = ("Anu", "Ravi", "John")
courses = ("Python", "SQL", "Power BI")
combined = []
for n, c in zip(names, courses):
        combined.append(n + "-" + c)
combined = tuple(combined)
print(combined)
```

```python
emails = ("anu@gmail.com", "ravi@yahoo.com", "john@edu.com")
domains = []
for email in emails:
        domain = email.split("@")[1]
        domains.append(domain)
domains = tuple(domains)
print(domains)
```
