```python
secret = 7

while True:
    guess = int(input("Enter number: "))

    if guess < secret:
        print("Too Low")
    elif guess > secret:
        print("Too High")
    else:
        print("You Won!")
        break
```

```python
num = int(input("Enter number: "))
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

print("Reversed:", rev)
```

```python
num = int(input("Enter number: "))
count = 0

while num > 0:
    num = num // 10
    count += 1

print("Digits:", count)
```

```python
sales = [1200, 800, 1500, 400, 2000]
count = 0

for sale in sales:
    if sale > 1000:
        print(sale)
        count += 1

print("Count:", count)
```

```python
total = 0

while True:
    num = int(input("Enter number (0 to stop): "))

    if num == 0:
        break

    total += num

print("Total sum:", total)
```
