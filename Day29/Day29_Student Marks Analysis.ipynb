```python
students = { 
    "Asha": {"Math": 78, "Science": 85, "English": 67}, 
    "Ravi": {"Math": 35, "Science": 40, "English": 50}, 
    "John": {"Math": 90, "Science": 92, "English": 88} 
}

results = {}
top_scorer = ""
highest_total = 0
failed_students = []

for name, marks in students.items():
    total = sum(marks.values())
    average = total / len(marks)
    
    results[name] = {"Total": total, "Average": average}
    
    if total > highest_total:
        highest_total = total
        top_scorer = name
    
    for mark in marks.values():
        if mark < 40:
            failed_students.append(name)
            break

print("Student Results:")
for name, data in results.items():
    print(name, "-> Total:", data["Total"], ", Average:", data["Average"])

print("\nTop Scorer:", top_scorer)
print("Failed Students:", failed_students)
```
