```python
# Task 1: Declare variables

order_id = "ORD12345"        # string
customer_name = "Rahul"      # string
product_price = 799.99       # float
quantity = 3                 # integer
discount = 0.10              # float (10%)
is_member = True             # boolean
```

```python
# Task 2: Print data types

print(type(order_id))
print(type(customer_name))
print(type(product_price))
print(type(quantity))
print(type(discount))
print(type(is_member))
```

```python
# Task 3: Total price before discount

total_price = product_price * quantity
print("Total Price before discount:", total_price)
```

```python
# Task 4: Apply discount

discount_amount = total_price * discount
price_after_discount = total_price - discount_amount

print("Price after discount:", price_after_discount)
```

```python
# Task 5: GST calculation

gst = 0.18
gst_amount = price_after_discount * gst
final_amount = price_after_discount + gst_amount

print("Final Amount including GST:", final_amount)
```

```python
# Task 6: Final Output

print("\n----- FINAL BILL -----")
print("Order ID:", order_id)
print("Customer:", customer_name)
print("Total Price:", total_price)

# Check if discount applied
if discount > 0:
    print("Discount Applied: Yes")
else:
    print("Discount Applied: No")

print("Final Amount (including GST):", round(final_amount, 2))
```
