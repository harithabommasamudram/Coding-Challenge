```python
trips = (
    (101, "Bangalore", 12.5, 250),
    (102, "Hyderabad", 8.0, 180),
    (103, "Bangalore", 15.2, 300),
    (104, "Chennai", 5.5, 120),
    (105, "Bangalore", 7.8, 200)
)
```

```python
total_trips = len(trips)
print("Total Trips:", total_trips)
```

```python
bangalore_count = 0

for trip in trips:
    if trip[1] == "Bangalore":
        bangalore_count += 1

print("Trips from Bangalore:", bangalore_count)
```

```python
total_revenue = 0

for trip in trips:
    total_revenue += trip[3]

print("Total Revenue:", total_revenue)
```

```python
print("High Distance Trips (>10 km):")

for trip in trips:
    if trip[2] > 10:
        print(trip)
```

```python
max_fare_trip = trips[0]

for trip in trips:
    if trip[3] > max_fare_trip[3]:
        max_fare_trip = trip

print("Highest Fare Trip:", max_fare_trip)
```

```python
total_fare = 0
count = 0

for trip in trips:
    if trip[1] == "Bangalore":
        total_fare += trip[3]
        count += 1

average_fare = total_fare / count
print("Average Fare (Bangalore):", average_fare)
```
