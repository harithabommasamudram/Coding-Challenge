```python
students = ("Anu", "Ravikumar", "Jo")

longest_name = students[0]

for name in students:
    if len(name) > len(longest_name):
        longest_name = name

print("Longest Name:", longest_name)
```

```python
subjects = ("Math", "Science", "English")

result = ", ".join(subjects)

print("Subjects:", result)
```

```python
students = ("Anu", "Ravi")

for name in students:
    reversed_name = name[::-1]
    print(reversed_name)
```

```python
students = ("Anu", "Ravi")

vowels = "aeiouAEIOU"

for name in students:
    count = 0
    for char in name:
        if char in vowels:
            count += 1
    print(name, "has", count, "vowels")
```

```python
courses = ("python", "sql", "power bi")

upper_courses = []

for course in courses:
    upper_courses.append(course.upper())

result = " | ".join(upper_courses)

print("Courses:", result)
```
